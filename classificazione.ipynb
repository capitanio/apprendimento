Prima di tutto importiamo le libreria che ci servono.

Si tratta di una delle librerie Python per machine learning più famose e usate.

```python
from numpy import where
from sklearn.datasets import make_blobs
```

Creiamo il  data set 1 di 1000 campioni con 2 blob di dati ben distinti.


```python


X1, y1 = make_blobs(n_samples=1000, centers=[[2,2], [7,7]], random_state=1)
```

Creiamo il  data set 2 di 1000 campioni con 2 blob di dati che si intersecano.


```python


X2, y2 = make_blobs(n_samples=1000, centers=[[2,2], [3,3]], random_state=1)
```

Il campione è composto da 2 caratteristiche (X array bidimensionale di 1000 righe e 2 colonne x1 e x2) e una etichetta y  che può essere 0 o 1 (es. maschio o femmina)

Adesso possiamo vedere un grafico del campione 1 (con blob ben distinti) 

```python
c=[(0, 500), (1, 500)]
from matplotlib import pyplot
# grafica il dataset assegnando il colore del generico punto in base alla sua classe di appartenenza (=valore di y)
for label, _ in c:
	row_ix = where(y1 == label)[0]
	pyplot.scatter(X1[row_ix, 0], X1[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```

Adesso possiamo vedere un grafico del campione 2 (con blob intersecantesi) 

```python
c=[(0, 500), (1, 500)]
from matplotlib import pyplot
# grafica il dataset assegnando il colore del generico punto in base alla sua classe di appartenenza (=valore di y)
for label, _ in c:
	row_ix = where(y2 == label)[0]
	pyplot.scatter(X2[row_ix, 0], X2[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```

Vediamo adesso come si comporterebbe un perceptrone con il dataset 1 (blob distinti)

```python

from sklearn.linear_model import Perceptron
classificatore = Perceptron(tol=1e-3, random_state=0)
classificatore.fit(X1, y1)
classificatore.score(X1, y1)
```

Vediamo adesso come si comporterebbe un perceptrone con il dataset 21 (blob intersecantesi)

```python

from sklearn.linear_model import Perceptron
classificatore = Perceptron(tol=1e-3, random_state=0)
classificatore.fit(X2, y2)
classificatore.score(X2, y2)
```
